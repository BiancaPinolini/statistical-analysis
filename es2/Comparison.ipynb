```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# number of iterations
N = 9
```

```python
with open('Direct_Sum.txt', 'r') as f:
    list_num_DS = f. read().splitlines()
```

```python
with open('Sorted_Sum.txt', 'r') as f:
    list_num_SS = f. read().splitlines()
```

```python
with open('Pairwise_Sum.txt', 'r') as f:
    list_num_PS = f. read().splitlines()
```

```python
with open('Kahan_Sum.txt', 'r') as f:
    list_num_KS = f. read().splitlines()
```

```python
F = np.arange(0,2*N-1,2)
D = np.arange(1,2*N,2)
```

```python
gammas_F_DS = []
gammas_LD_DS = []

for i in F:
    gammas_F_DS.append(abs(float(list_num_DS[i])))
    
for j in D:
    gammas_LD_DS.append(abs(float(list_num_DS[j])))
```

```python
gammas_F_SS = []
gammas_LD_SS = []

for i in F:
    gammas_F_SS.append(abs(float(list_num_SS[i])))
    
for j in D:
    gammas_LD_SS.append(abs(float(list_num_SS[j])))
```

```python
gammas_F_PS = []
gammas_LD_PS = []

for i in F:
    gammas_F_PS.append(abs(float(list_num_PS[i])))
    
for j in D:
    gammas_LD_PS.append(abs(float(list_num_PS[j])))
```

```python
gammas_F_KS = []
gammas_LD_KS = []

for i in F:
    gammas_F_KS.append(abs(float(list_num_KS[i])))
    
for j in D:
    gammas_LD_KS.append(abs(float(list_num_KS[j])))
```

```python
x = np.arange(1,N+1)

x_name = []

for i in range(1,N+1):
    x_name.append(f'$10^{i}$')
```

# FLOAT

```python
plt.figure(figsize=(10,7))
plt.plot(x, gammas_F_DS, color='blue', marker='o', label='Direct Sum')
plt.plot(x, gammas_F_SS, color='magenta', marker='o', label='Sorted Sum')
plt.plot(x, gammas_F_PS, color='green', marker='o', label='Pairwise Sum')
plt.plot(x, gammas_F_KS, color='red', marker='o', label='Kahan Sum')
plt.xticks(x, x_name, fontsize=16)
plt.legend(fontsize=16)
plt.xlabel('Number of Iterations', fontsize=16)
plt.ylabel('$\epsilon_{alg}$', fontsize=16)
plt.yscale('log')
plt.title("Computation of $\epsilon_{alg}$ with Float numbers", fontsize=16)
plt.show()
```

# LONG DOUBLE

```python
plt.figure(figsize=(10,7))
plt.plot(x, gammas_LD_DS, color='blue', marker='o', label='Direct Sum')
plt.plot(x, gammas_LD_SS, color='magenta', marker='o', label='Sorted Sum')
plt.plot(x, gammas_LD_PS, color='green', marker='o', label='Pairwise Sum')
plt.plot(x, gammas_LD_KS, color='red', marker='o', label='Kahan Sum')
plt.xticks(x, x_name, fontsize=16)
plt.legend(fontsize=16)
plt.xlabel('Number of Iterations', fontsize=16)
plt.ylabel('$\epsilon_{alg}$', fontsize=16)
plt.yscale('log')
plt.title("Computation of $\epsilon_{alg}$ with Long Double numbers", fontsize=16)
plt.show()
```
