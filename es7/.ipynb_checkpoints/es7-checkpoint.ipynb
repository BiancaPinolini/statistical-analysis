```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu_s = [0,0]
cov_s = [[0.09, 0.045], [0.045, 0.09]]
```

```python
mu_b = [4,4]
cov_b = [[1., 0.4], [0.4, 1.]]
```

```python
x_s, y_s = np.random.multivariate_normal(mu_s, cov_s, 50000).T
x_b, y_b = np.random.multivariate_normal(mu_b, cov_b, 50000).T
```

```python
plt.figure(figsize=(7, 7))
plt.plot(x_s, y_s, '.', color = 'red', label = 'signal')
plt.plot(x_b, y_b, '.', color = 'blue', label = 'background')
plt.legend()
plt.show()
```

```python
plt.hist(x_s, bins=200, color = 'red', histtype = 'step', label = 'signal')
plt.hist(x_b, bins=200, color = 'blue', histtype = 'step', label = 'background')
plt.xlabel('X')
plt.legend()
plt.show()
```

```python
plt.hist(y_s, bins=200, color = 'red', histtype = 'step', label = 'signal')
plt.hist(y_b, bins=200, color = 'blue', histtype = 'step', label = 'background')
plt.xlabel('Y')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
import sklearn.datasets
import sklearn.linear_model
import matplotlib
from sklearn.preprocessing import StandardScaler, LabelBinarizer
```

```python
# keras imports
import tensorflow.keras.models as Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Dropout, Flatten
from tensorflow.keras.utils import plot_model
from tensorflow.keras import backend as K
from tensorflow.keras import metrics
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, TerminateOnNaN
```

```python
s_b = np.array([])
```

```python
for i in range (50000):
    s_b = np.append(s_b,1)
```

```python
for i in range (50000):
    s_b = np.append(s_b,0)
```

```python
data = pd.DataFrame(
    {
        'x': x,
        'y': y,
        's_b': s_b
    }
)
```

```python
data.head()
```

# let's start

```python
dropoutRate = 0.25
```

```python
X = data[['x', 'y']]
y = data['s_b']
```

```python
X = StandardScaler().fit_transform(X)
y = LabelBinarizer().fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = Sequential()
```

```python
model.add(Dense(units=8, input_dim=2, activation='relu'))
model.add(Dense(units=8, activation='relu'))
model.add(Dropout(dropoutRate))
model.add(Dense(units=2, activation='softmax'))
model.add(Dropout(dropoutRate))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', min_delta=1e-15, patience=10, verbose=1, mode='auto', baseline=None)
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), verbose=1, epochs = 120, shuffle=True, callbacks=[early_stop])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss function')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['Train','Test'])
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Accuracy")
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['train', 'test'])
plt.show()
```
